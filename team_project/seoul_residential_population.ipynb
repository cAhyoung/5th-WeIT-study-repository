```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rc('font', family='NanumBarunGothic') 
plt.rcParams['axes.unicode_minus'] =False
```

```python
data = pd.read_excel("서울시_구별_가구원수_통계.xlsx", sheet_name="data_transpose")
data
```

```python
### 구별 가구수만 추출
tot_rsd_pop = data[data["동별(2).1"] == "계 (가구)"]
tot_rsd_pop = tot_rsd_pop.T
tot_rsd_pop.columns = [2018, 2019, 2020, 2021, 2022, 2023]
tot_rsd_pop.drop(["동별(2)", "동별(2).1"], inplace=True)
tot_rsd_pop
```

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

def extracting_data(option):
  extract_rsd_pop = data[data["동별(2).1"] == option]
  extract_rsd_pop = extract_rsd_pop.T
  extract_rsd_pop.columns = [2018, 2019, 2020, 2021, 2022, 2023]
  extract_rsd_pop.drop(["동별(2)", "동별(2).1", "소계"], inplace=True)
  
  return extract_rsd_pop

def vis_lineplot(data, title):
  fig = go.Figure()

  # 각 지역에 대해 선 그래프 추가
  for idx, row in data.iterrows():
      fig.add_trace(go.Scatter(x=data.columns, y=row, mode='lines+markers', name=idx))
      
  # 그래프 레이아웃 설정
  fig.update_layout(
      title=title,
      xaxis_title='년도',
      yaxis_title='전체 가구 수 (가구)',
      xaxis=dict(tickmode='linear'),  # x축 눈금 조정
      legend_title='지역',
      template='plotly_white'
  )

  # 그래프 출력
  fig.show()
```

```python
data["동별(2).1"].unique()
```

```python
for option in data["동별(2).1"].unique():
  extracted_data = extracting_data(option)
  vis_lineplot(extracted_data, f"{option} 추이")
```

```python
data2 = data.copy()
data2
```

```python
data2 = data2[data2["동별(2)"] == 2022]
data2
```

```python
data2 = data2[(data2["동별(2).1"] == "1인가구 (가구)") | (data2["동별(2).1"] == "2인가구 (가구)") | (data2["동별(2).1"] == "3인가구 (가구)") | (data2["동별(2).1"] == "4인가구 (가구)")]
data2
```

```python
data2 = data2.T
data2.columns = ["1인가구 수", "2인가구 수", "3인가구 수", "4인가구 수"]
```

```python
data2.drop(["동별(2)", "동별(2).1"], inplace=True)
```

```python
data2 = data2.loc[["종로구", "중구", "강북구", "관악구"], :]
data2
```

```python
data2["3~4인가구 수"] = data2["3인가구 수"] + data2["4인가구 수"]
data2
```

```python
data2.drop(["3인가구 수", "4인가구 수"], axis=1, inplace=True)
```

```python
data2
```

```python
# 인덱스를 열로 변환
data2.reset_index(inplace=True)
data2.rename(columns={'index': '구'}, inplace=True)
```

```python
data2
```

```python
# 데이터 변환: 넓은 형식에서 긴 형식으로 변환
df_melted = data2.melt(id_vars='구', var_name='가구 유형', value_name='가구 수')

# Plotly로 시각화
fig = px.bar(
    df_melted,
    x='가구 유형',
    y='가구 수',
    color='구',
    barmode='group',
    labels={'가구 유형': '가구 유형', '가구 수': '가구 수'},
    title='1인, 2인, 3~4인 가구 유형별 가구 수 비교'
)
```

```python
fig.show()
```
