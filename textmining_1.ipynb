```python
### 필요한 패키지 import
import pandas as pd
import numpy as np
import seaborn as sns
import re
import konlpy
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
plt.rc('font', family='NanumBarunGothic') 
plt.rcParams['axes.unicode_minus'] =False
```

```python
moon = open("speech_moon.txt", encoding="UTF-8").read()
moon
```

```python
### 한글이 아닌 모든 문자는 제거
moon = re.sub('[^가-힣]', ' ', moon)
moon
```

```python
hannanum = konlpy.tag.Hannanum()

hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
nouns =hannanum.nouns(moon)
nouns
```

```python
df_word = pd.DataFrame({'word' : nouns})
df_word
```

```python
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
df_word2 =df_word.groupby('word', as_index=False) \
                .agg(n=('word', 'count')) \
                .sort_values('n', ascending=False)

df_word2
```

```python
top20 = df_word2.head(20)
top20
```

```python
sns.barplot(data=top20, y='word', x='n')
```

```python
### 워드클라우드 만들기
dic_word = df_word2.set_index('word').to_dict()['n']
dic_word
```

```python
font = "NanumGothic-Regular.ttf"
```

```python
wc = WordCloud(random_state=42, 
               width=600, 
               height=500, 
               background_color="white", 
               font_path=font)

img_wc = wc.generate_from_frequencies(dic_word)

plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(img_wc)
```
